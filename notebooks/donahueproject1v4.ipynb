```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from numpy import linalg as la
from scipy.optimize import curve_fit
```

```python
df = pd.read_csv('../data/New Data 10-22.csv')
x = np.array(df['Hours_Studied'])
y = np.array(df['Exam_Score'])
```

```python
for x,y in zip(x,y):
    print("If a student studies %s hours they would earn a %0.0f as an exam score"%(x,y))
```

```python
fig,ax = plt.subplots();
scatt_plot = ax.plot(x,y,"ro");
ax.set_xlim(8, 30)
ax.set_ylim(29,75)
ax.set_title("Hours studied vs exam score");
ax.set_xlabel('Hours Studied');
ax.set_ylabel('Exam Score');
```

```python
def arb_linear(x,a,b): return a*x+b
def arb_quad(x,a,b,c): return a*x**2+b*x+c
```

```python
lcoef,lcovar=curve_fit(arb_linear,x,y)
qcoef,qcovar=curve_fit(arb_quad,x,y)
```
